```python
import pandas as pd
import numpy as np

feature_nums = [1,5,10,13,16,17,18,19,21,25]

data = pd.read_csv('..\\Fermi-LAT Data\\fl_varranked_rm0.csv')

features = list(np.zeros(len(feature_nums)))

for i in range(0,len(features)):
    features[i] = data.columns[feature_nums[i]]

print(features)
```

```python
newdata = pd.DataFrame()
for i in range(0,len(features)):
    newdata[features[i]]=data[features[i]]
print(newdata)
newdata['CLASS1'] = data['CLASS1']
newdata.to_csv('..\\Fermi-LAT Data\\fl_wrappered.csv')
newdata.head()
```
