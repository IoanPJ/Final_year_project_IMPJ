```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame()

for i in range(1,11):
    df = pd.read_csv('..\\..\\Search Area Results\\iteration_'+str(i)+'_raw.csv')
    data[f'iteration_{i}']=df.loc[df['CLASS1']==3,'y_probs_bll']


data['mean'] = data.mean(axis=1)
data=data.sort_values(by='mean')
means=np.array(data['mean'])
data=data.drop('mean',axis=1)
stderrs = data.sem(axis=1)
stderrs=np.array(stderrs)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax = sns.swarmplot(data=means,orient='h')
```
