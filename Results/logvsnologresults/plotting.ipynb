```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os.path import basename, exists
import matplotlib as mpl

def download(url):
    filename = basename(url)
    if not exists(filename):
        from urllib.request import urlretrieve
        local, _ = urlretrieve(url, filename)
        print('Downloaded ' + local)
        
download('https://github.com/AllenDowney/AstronomicalData/raw/main/' + 'az-paper-twocol.mplstyle')
plt.style.use('./az-paper-twocol.mplstyle')
```

```python
log = pd.read_csv('logrocresults.csv')
nolog=pd.read_csv('nologrocresults.csv')


fig, ax = plt.subplots(figsize=(4,4),dpi=300)
n=14
ax.plot(log['FPR'],log['TPR'],color='blue',linestyle=':',marker='v',label='Log Transformed')
ax.plot(nolog['FPR'],nolog['TPR'],linestyle=':',color='red',marker='x',label='Non Log Transformed')
ax.set_ylabel('True Positive Rate',fontsize=n)
plt.xticks(fontsize=n)
plt.yticks(fontsize=n)
ax.set_xlabel('False Positive Rate',fontsize=n)
ax.legend(fontsize=11,loc='lower right')
```
